Let one side have length $L$ where $0<L\le 1$.  Let the angles be $\theta$ and $\pi-\theta$,
where $0<\theta\le\pi/2$.  That means the other side has length $1/(L\sin\theta)$.

Using the side of length $L$ as the circumference, the height is $1/L$, the area of the
base is $L^2/4\pi$, and the volume is $L/4\pi$.

Using the side of length $1/L\sin\theta$ as the circumference, the height is $L\sin\theta$,
the base is $1/4\pi L^2\sin^2\theta$, and the volume is $1/4\pi L\sin\theta$.

$$ V = \frac{1}{2}\left(\frac{L}{4\pi} + \frac{1}{4\pi L\sin\theta}\right) $$

Since $0<\theta\le\pi/2$, $0<\sin\theta\le1$, so $V$ is minimized when $\sin\theta = 1$,
or $\theta=\pi/2$.

Then,

$$ \frac{dV}{dL} = \frac{1}{8\pi}\left(1 - \frac{1}{L^2}\right) $$

so $V$ is minimized when $L = 1$, so the minimum $V$ is $1/4\pi$, which, unsurprisingly,
is when the parallelogram is a square.
