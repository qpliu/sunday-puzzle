Fiddler
-------
Due to symmetry, only consider a quarter-circle.

Let the radius be $1$ and let $x = 0$, $y = 0$ be the center of Diametric Beach.

The distances of point $(x,y)$ are $y$ and $1 - \sqrt{x^2+y^2}$.

Let $y=f(x)$ be the equidistant points.  The probability of being closer to Diametric Beach
is then $\frac4\pi\int_0^1 dx\,f(x)$.

For the equidistant points,

$$\begin{aligned}
    f &= 1 - \sqrt{x^2+f^2} \\
    x^2 + f^2 &= (1-f)^2 \\
    x^2 &= 1 - 2f \\
    f &= (1 - x^2)/2 \\
\end{aligned}$$

```python
plot([sqrt(1-x^2), (1-x^2)/2], (0,1))
```

```python
p = integral(4/pi*(1-x^2)/2, x, 0, 1)
(p, numerical_approx(p))
```

Extra credit
------------
Integrating over the quarter circle, the average distance is

$$ \frac4\pi\int_0^1 dx \left( \int_0^{(1-x^2)/2} dy\,y + \int_{(1-x^2)/2}^{\sqrt{1-x^2}} dy\,(1 - \sqrt{x^2+y^2}) \right) $$

```python
y = var('y')
d = 4/pi*integral(integral(y, y, 0, (1-x^2)/2) + integral(1-sqrt(x^2+y^2),y,(1-x^2)/2,sqrt(1-x^2)), x, 0, 1)
(d, numerical_approx(d))
```
