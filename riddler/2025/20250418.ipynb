[2025-04-18 Fiddler](https://thefiddler.substack.com/p/can-you-throw-the-hammer)
====================

I'll assume reaching 4 points on a 2-point hole also wins.

I'm having a hard time imagining when it would be advantageous
to reject the hammer.  If you are up 2-0, it might be, but it's
not obvious without working through the probabilities.

Fiddler
-------
If the score is 2-2, either player has a 50% chance of winning,
irrespective of the hammer.

If the score is 2-1, the losing player will never reject the
hammer.  If the winning player rejects the hammer, the score
will be 2-2, otherwise either player has a 50% chance of winning.
If no hammer is thrown, the winning player will have a 75%
chance of winning, so the losing player will always throw the
hammer.  So a 2-1 score means either player has a 50% chance of
winning.

If the score is 2-0, the losing player will never reject the
hammer.  If the winning player rejects the hammer, it will be
2-1 and each player has a 50% chance of winning, so the winning
player will not reject the hammer, since accepting the hammer
means having a 75% chance of winning.  If no hammer is thrown,
the winning player will have a 75% chance of winning.  So a
2-0 score means the winning player has a 75% chance of winning.

If the score is 1-1, a rejected hammer means going to 2-1 and
a 50% chance of winning.  An accepted hammer means a 50% chance
of winning.  No hammer means going to 2-1 and a 50% chance of
winning.  So a 1-1 score means a 50% chance of winning.

Finally, if the score is 1-0 and no hammer is thrown, the winning
player has 5/8 chance of winning.  If the winning player rejects
the hammer, it's 50%.  If the losing player rejects the hammer,
that means 25%.  If the hammer is accepted, the winning player
has 75% chance of winning.  That means the winning player will
throw the hammer and have a 75% chance of winning irrespective
of whether the losing player accepts the hammer.

Extra credit
------------
4-4 means a 50% chance of winning.

4-3 means the losing player will throw the hammer for a 50% chance of winning.

At 4-2, if no hammer is thrown, the winning player has a 75% chance of winning.
If the hammer is accepted, the winning player has a 75% chance of winning.  The
losing player will not reject the hammer for a 0% chance of winning.  The winning
player will not reject the hammer for a 50% chance of winning.

At 4-1, if no hammer is thrown, the winning player has a 7/8 chance of winning.
If the hammer is accepted, the winning player has a 75% chance of winning.  The
losing player will not reject the hammer for a 0% chance of winning.  If the winning
player rejects the hammer, it'll be a 75% chance of winning.  So the losing player
will always throw the hammer for a 25% chance of winning.

At 4-0, if no hammer is thrown, the winning player has a 7/8 chance of winning.
If the hammer is accepted, the winning player has 7/8 chance of winning.  The
losing player will not reject the hammer for 0%, and the winning player will not
reject the hammer for 75%.

3-3 means a 50% chance of winning.

At 3-2, if no hammer is thrown, the winning player has a 7/8 chance of winning.
If the hammer is accepted, the winning player has a 75% chance of winning.
The winning player will not reject the hammer for 50% from 3-3.  If the losing player
rejects the hammer, the winning player has a 75% chance of winning from 4-2.  So
the losing player will always throw the hammer for a 25% chance of winning.

At 3-1, if no hammer is thrown, the winning player has a 75% chance of winning.
If the hammer is accepted, the winning player has a 75% chance of winning.
The winning player can reject the hammer for a 75% chance of winning from 3-2.  The
losing player can reject the hammer for a 25% chance of winning from 4-1.

At 3-0, if no hammer is thrown, the winning player has a 13/16 chance of winning.
If the hammer is accepted, the winning player has a 7/8 chance of winning.  The losing
player can reject the hammer for a 1/8 chance of winning from 4-0.  So the winning
player will always throw the hammer for a 7/8 chance of winning.

2-2 means a 50% chance of winning.

At 2-1, if no hammer is thrown, the winning player has a 5/8 chance of winning.
If the hammer is accepted, the winning player has a 50% chance of winning.
The winning player can reject the hammer for 50% from 2-2.  The losing player
will never reject the hammer for 25% from 3-1.  So the losing player will always
throw the hammer for 50%.

At 2-0, if no hammer is thrown, the winning player has 11/16.  If the hammer is
accepted, the winning player has 11/16.  The winning player will not reject
the hammer for 50% from 2-1.  The losing player will not reject the hammer for
3/16 from 3-0.

1-1 means 50%.

At 1-0, if no hammer is thrown, the winning player has 19/32.  If the hammer is
accepted, the winning player has 11/16.  The winning player will not reject
the hammer for 50% from 1-1.  The losing player can reject the hammer for 5/16
from 2-0.  So the winning player will always throw the hammer for for 11/16.
