Fiddler
-------
When $A < 3$, it's always possible to put the unit square in the center
of the square, with the squares not rotated with respect to each other.

First, leaving out the corners, it is not possible to fit a unit square
between two parallel lines separated by $(A - 1)/2 < 1$.

When $A < 2 + 1/\sqrt2$, it's not possible to even put a circle with
diameter $1$ in the corners.

When $A \ge 3$, it should always be possible to put the second unit square
in a corner.  Consider putting a circle with diameter $\sqrt2$, which
circumscribes a unit square, anywhere within the larger square.  It is not
possible for the circle to cover a portion of each of the 4 unit squares in
the 4 corners.

Consider a unit square with corners at $(a,0)$ and $(0,b)$, where $a < 1$
and $b = \sqrt{1-a^2}$.  The other two corners are at $(b,a+b)$ and $(a+b,a)$.
Let $(c,c)$ be where the unit square borders the corner of the other unit square.

```python
a = 0.25
b = sqrt(1-a^2)
c = solve((x-b)*(x-a) == (x-a-b)^2, x)[0].rhs()
(polygon([(a,0),(a+b,a),(b,a+b),(0,b)],alpha=0.25)
 + polygon([(0,0),(a+b,0),(a+b,a+b),(0,a+b)],fill=false,linestyle='dotted')
 + line([(0,0),(a+b,a+b)],linestyle='dashed')
 + line([(c,a+b),(c,c),(a+b,c)],color='black')
 + polygon([(c,a+b),(c,c),(a+b,c),(a+b,a+b)],color='black',alpha=0.25)
)
```

```python
a = var("a")
b = sqrt(1-a^2)
c = solve((x-a)*(x-b) == (x-a-b)^2, x)[0].rhs()
print(c)
print(solve(derivative(c, a) == 0, a))
plot(c, a, 0, 1)
```

So it's not possible for $c < 1$, and $c = 1$ corresponds to $A = 3$,
so the minimum $A = 3$.

Extra credit
------------
With 3 unit squares, the corners still have to be covered, so when $B$ is
less than the minimum, at least one square will have to cover parts of two
of the squares in two of the corners.  If opposite corners are covered by
one unit square, then $B < 3$, as in the Fiddler answer.  But I suspect that
the minimum $B$ is larger than that when adjacent corners are covered by
one unit square.

Let the two adjacent unit squares be at $(0,0)$ and $(B-1,0)$.  If
$B \ge 2 + \sqrt2$, a single unit square cannot partially cover both corner
unit squares, so an upper bound on the minimum $B$ is $2+\sqrt2$.

And, from the $c \ge 1$ calculation in the Fiddler answer, if a unit square
covers both $(1,1)$ and $(B-1,1)$, a unit square in the corner cannot be rotated
to avoid the covered part without leaving the larger square.

So put one unit square centered at $(B/2,1)$ and rotated 45.  There is no
room between the square and the side from $(0,0)$ to $(B,0)$ for a unit square.

The other two unit squares must be arranged occupy enough of the other side to
crowd out the fourth unit square.  One must cover $(1,B-1)$ and the other must
cover $(B-1,B-1)$ to prevent squares in the corners.  They must be close enough
to each other to prevent squares between them, as well as being close enough to
the first unit square.  The figure shows that there are many possible ways to
place these two unit squares when $B < 2+\sqrt2$ to obstruct a fourth unit square.

```python
b = 2+sqrt(2)
(polygon([(0,0),(b,0),(b,b),(0,b)],fill=false)
 + polygon([(1,1),(b/2,1-1/sqrt(2)),(b-1,1),(b/2,1+1/sqrt(2))],alpha=0.25)
 + polygon([(0,0),(1,0),(1,1),(0,1)],alpha=0.25,color='red')
 + polygon([(b,0),(b-1,0),(b-1,1),(b,1)],alpha=0.25,color='red')
 + polygon([(0,sqrt(2)),(1/sqrt(2),1/sqrt(2)),(2/sqrt(2),2/sqrt(2)),(1/sqrt(2),sqrt(2)+1/sqrt(2))],alpha=0.25,color='red')
 + polygon([(b,sqrt(2)),(b-1/sqrt(2),1/sqrt(2)),(b-2/sqrt(2),2/sqrt(2)),(b-1/sqrt(2),sqrt(2)+1/sqrt(2))],alpha=0.25,color='red')
 + polygon([(1/2,2+1/sqrt(2)),(3/2,2+1/sqrt(2)),(3/2,1+1/sqrt(2)),(1/2,1+1/sqrt(2))],alpha=0.25)
 + polygon([(b-1/2,2+1/sqrt(2)),(b-3/2,2+1/sqrt(2)),(b-3/2,1+1/sqrt(2)),(b-1/2,1+1/sqrt(2))],alpha=0.25)
)
```
