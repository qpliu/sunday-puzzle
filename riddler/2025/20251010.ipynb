Fiddler
-------
The probability of a single roll is

```python
r1(n) = (6 - abs(7-n))/36
```

The probability of a set of three different rolls is

```python
r3(i,j,k) = r1(i)*r1(j)*r1(k)*factorial(3)
```

The sum of the probabilities of all winning rolls is

```python
p3 = (r3(3,4,5) + r3(6,7,8) + r3(9,10,11)
      + r3(3,6,9) + r3(4,7,10) + r3(5,8,11)
      + r3(3,7,11) + r3(5,7,9))
p3,numerical_approx(p3)
```

And the answer is 113/1944, about 5.8%.

Extra credit
------------
Using code, the answer is 910405/2519424, about 36%.

```python
from functools import cache

winners = [frozenset([3,4,5]),frozenset([6,7,8]),frozenset([9,10,11]),
           frozenset([3,6,9]),frozenset([4,7,10]),frozenset([5,8,11]),
           frozenset([3,7,11]),frozenset([5,7,9])]

@cache
def pn(grid,nrolls):
    if nrolls <= 0:
        return 0
    p = 0
    for r in [2..12]:
        pr = r1(r)
        ngrid = grid | frozenset([r])
        if len([w for w in winners if ngrid >= w]) > 0:
            p += pr
        else:
            p += pr*pn(ngrid,nrolls-1)
    return p
```

```python
[(n, pn(frozenset(), n), numerical_approx(pn(frozenset(), n))) for n in [3..15]]
```
