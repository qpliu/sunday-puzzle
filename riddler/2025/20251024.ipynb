Fiddler
-------
Each dollar bet on Judge can lose the oddsmaker 2/3 dollar.  Each dollar bet on Raleigh can lose the
oddsmaker 11/10 dollar.

If Judge wins, the oddsmaker's take is $1-f-2/3\cdot f = 1 - 4/3\cdot f$.

If Raleigh wins, the oddsmaker's take is $f - 11/10\cdot(1-f) = -11/10 + 21/10\cdot f$.

When they are equal $f = 63/113$, and the oddsmaker gets $8/113$ for each dollar bet.

```python
s = solve(-2/3*x + 1-x == x - 11/10*(1-x), x)
f = s[0].rhs()
s, -2/3*f + 1-f, f - 11/10*(1-f)
```

### Part 1
If A wins, the oddsmaker takes $1-f - fx$ per dollar bet.

If B wins and $1-f$ is bet on B, the maximum the oddsmaker can take is $f$.
If less than that is bet on B, the oddsmaker can take up to $1$, depending
on how much is bet on B, so it doesn't make sense to consider the possibility
that less than $1-f$ is bet on B.

Then, for them to be equal, $f = 1/(2+x)$, below which the oddsmaker cannot
take as much when B wins as when A wins no matter the odds.

### Part 2
If A wins, the oddsmaker takes $1-f - f/y$ per dollar bet.

By the same reasoning as in part 1, the minimum $f$ is $1/(2+1/y) = y/(2y+1)$.
