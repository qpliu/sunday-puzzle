Fiddler
-------
There are 5!=120 possible drawings, assuming that drawings continue until everyone has drawn
before starting over if necessary.

Assuming A draws E, there are 24 possible drawings, which is small enough to enumerate by hand.

* EABCD - single loop
* ~~EABDC~~ - D draws D
* ~~EACBD~~ - C draws C
* ~~EACDB~~ - C draws C
* EADBC - single loop
* EADCB - two loops
* ~~EBACD~~ - B draws B
* ~~EBADC~~ - B draws B, D draws D
* ~~EBCAD~~ - B draws B, C draws C
* ~~EBCDA~~ - B draws B, C draws C, D draws D
* ~~EBDCA~~ - B draws B
* ~~EBDAC~~ - B draws B
* ECABD - single loop
* ~~ECADB~~ - D draws D
* ECBAD - two loops
* ~~ECBDA~~ - D draws D
* ECDAB - single loop
* ECDBA - two loops
* EDABC - two loops
* EDACB - single loop
* EDBAC - single loop
* EDBCA - two loops
* ~~EDCAB~~ - C draws C
* ~~EDCBA~~ - C draws C

There are 11 possible drawings that don't start over, and 6 of them have a single loop,
so the probability of a single loop is 6/11.

Extra credit
------------
For $N$ students, the number of possible drawings that don't start over is $!N$,
and the number of possible drawings that form a single loop is $(N-1)!$, so the
probability drawing a single loop is $(N-1)!/!N = (N-1)!/[\frac{N!}{e}]$, which
is 6/11 when $N = 5$, and, when $N$ is large, is approximately $e/N$.
