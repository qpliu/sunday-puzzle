Fiddler
-------
Let the pivot point be at $x = 0$, $y = 0$.  The center of mass is

```python
x,y,L = var("x,y,L")

m = integral(integral(1,y,0,2), x, 0, 1) + integral(integral(1,y,1,2), x, -L, 0)
xc = (integral(integral(x,y,0,2), x, 0, 1) + integral(integral(x,y,1,2), x, -L, 0))/m
yc = (integral(integral(y,y,0,2), x, 0, 1) + integral(integral(y,y,1,2), x, -L, 0))/m
xc, yc
```

```python
plot([xc, yc], L, 0, 2)
```

```python
solve(xc == 0, L)
```

And the tower tips over when $L > \sqrt2$.

Extra credit
------------
Doing a similar calculation, where the pivot point is
at $x = 1$, $y = 0$, or $r = 1$, $\theta = 0$,

```python
r,theta,alpha = var('r,theta,alpha')
assume(theta > 0, theta < pi)
m = integral(integral(r,r,1,2),alpha,0,theta)
xc = integral(integral(r*r*cos(alpha),r,1,2),alpha,0,theta)/m
yc = integral(integral(r*r*sin(alpha),r,1,2),alpha,0,theta)/m
xc, yc
```

```python
plot([xc-1,yc], theta, 0, 3*pi/4)
```

```python
solve(xc == 1, theta)
```

```python
theta_c = find_root(xc == 1, 0, 3*pi/4)
theta_c,numerical_approx(theta_c/pi),numerical_approx(theta_c/pi*180)
```

```python
(line([(1,0),(2,0)])
 + line([(cos(theta_c),sin(theta_c)),(2*cos(theta_c),2*sin(theta_c))])
 + arc((0,0),1,sector=(0,theta_c))
 + arc((0,0),2,sector=(0,theta_c))
)
```

And the tower tips over when $\theta \gtrsim 1.55537 \approx 89.116\degree$.
