Fiddler
-------
The amounts must be the same.

The total amount of coffee must add up to 12 and the total amount of tea must add up to 12,
so if the coffee cup has $t$ fl oz of tea, it must have $12-t$ fl oz of coffee, leaving
$12-t$ fl oz of tea $12 - (12-t) = t$ fl oz of coffee for the tea cup.

Extra credit
------------
The most tea that can be in the coffee cup after doing all the steps is 6 fl oz.

Let $f$ be the fraction of tea in the tea cup, which must be the same as the fraction
of coffee in the coffee cup.

If $12t$ fl oz of liquid is poured from the tea cup, then the coffee will contain
$12f + 12t(1-f)$ fl oz of coffee.

After mixing, the fraction of coffee in the coffee cup will be
$\frac1{1+t}f + \frac{t}{1+t}(1-f)$, which is the weighted average of $f$ and $1-f$.

The weight of $1 - f$ cannot be more than $1/2$, since $0 \le t \le 1$, so if $f > 1/2$,
the mixing can never make $f$ less than $1/2$.

Since $f = 1$ initially, $f$ can never be less than $1/2$.
