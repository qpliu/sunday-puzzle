Fiddler

The amounts must be the same.

The total amount of coffee must add up to 12 and the total amount of tea must add up to 12,
so if the coffee cup has $t$ fl oz of tea, it must have $12-t$ fl oz of coffee, leaving
$12-t$ fl oz of tea $12 - (12-t) = t$ fl oz of coffee for the tea cup.

Extra credit
------------
Let $f$ be the fraction of coffee in the coffee cup.

If $12t$ fl oz of tea is added and $12t$ fl oz of the new mixture is removed,
the fraction of coffee in the coffee cup becomes $\frac{f}{1+t}$.

If $t = 1$, the final fraction is $1/2$.

If $t = 1/2$, the process can be done twice, taking the fraction from $1$ to $2/3$,
then to the final fraction of $4/9$.

If $t = 1/n$, the process can be done $n$ times, the final fraction is
$\left(\frac{n}{n+1}\right)^n$.
