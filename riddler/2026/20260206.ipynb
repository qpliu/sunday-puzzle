[2026-02-06 Fiddler](https://thefiddler.substack.com/p/how-long-is-the-all-star-streak)
====================

Fiddler
-------
The probability that the streak is 1 is 1/2.
The probability that the streak is 2 is 1/4.
The probability that the streak is 3 is 1/8.
And so on.

```python
sum(x/2^x,x,1,oo)
```

So the average streak is 2.

Extra credit
------------
Let $p(a,b,c)$ be the probability that $a$, $b$, $c$ are the current streak, and previous game was
$b$ versus $c$ and the game about to be played is $a$ versus $b$.

One of $b$ and $c$ must be zero and the other must be non-zero.

At most one of $a$, $b$, and $c$ can be greater than one.

Enumerating the states and their probabilities, where $n \ge 3$,

$$\begin{aligned}
    p(0,0,1) &= \frac12 \sum_k p(k,0,0) = 0 \\
    p(0,0,2) &= \frac12 \sum_k p(k,1,0) \\
    p(0,0,n) &= \frac12 p(0,n-1,0) + \frac12 p(1,n-1,0) \\
    p(0,1,0) &= \frac12 \sum_k p(0,k,0) \\
    p(0,2,0) &= \frac12 \sum_k p(1,k,0) \\
    p(0,n,0) &= \frac12 p(n-1,0,0) + \frac12 p(n-1,1,0) \\
    p(1,0,1) &= \frac12 \sum_k p(k,0,1) \\
    p(1,0,2) &= \frac12 p(0,1,1) = 0 \\
    p(1,0,n) &= \frac12 p(0,n-1,1) = 0 \\
    p(1,1,0) &= \frac12 \sum_k p(0,k,1) = \frac12 p(0,0,1) = 0 \\
    p(1,2,0) &= \frac12 p(1,0,1) \\
    p(1,n,0) &= \frac12 p(n-1,0,1) \\
    p(2,0,1) &= \frac12 p(0,0,2) + \frac12 p(1,0,2) \\
    p(2,1,0) &= \frac12 p(0,0,2) \\
    p(n,0,1) &= \frac12 p(0,0,n) + \frac12 p(1,0,n) = \frac12 p(0,0,n) \\
    p(n,1,0) &= \frac12 p(0,0,n) + \frac12 p(0,1,n) = \frac12 p(0,0,n) \\
\end{aligned}$$

Based on numerical simulations, hypothesize $p(0,0,k) = 2^{-(k+1)}$ for $k \ge 2$,
$p(0,k,0) = 2^{-(k+2)}$ for $k \ge 1$,
$p(1,0,1) = 1/8$, $p(1,k,0) = p(k,0,1) = p(k,1,0) = 2^{-(k+2)}$ for $k \ge 2$.

$$\begin{aligned}
    p(0,0,1) &= 0 \\
    p(0,0,2) &= \frac12 \sum_k p(k,1,0) = \frac18 \\
    p(0,0,n) &= \frac12 p(0,n-1,0) + \frac12 p(1,n-1,0) = 2^{-(n+1)} \\
    p(0,1,0) &= \frac12 \sum_k p(0,k,0) = \frac18 \\
    p(0,2,0) &= \frac12 \sum_k p(1,k,0) = \frac{1}{16} \\
    p(0,n,0) &= \frac12 p(n-1,0,0) + \frac12 p(n-1,1,0) = 2^{-(n+2)} \\
    p(1,0,1) &= \frac12 \sum_k p(k,0,1) = \frac18 \\
    p(1,0,2) &= \frac12 p(0,1,1) = 0 \\
    p(1,0,n) &= \frac12 p(0,n-1,1) = 0 \\
    p(1,1,0) &= \frac12 \sum_k p(0,k,1) = \frac12 p(0,0,1) = 0 \\
    p(1,2,0) &= \frac12 p(1,0,1) = \frac{1}{16} \\
    p(1,n,0) &= \frac12 p(n-1,0,1) = 2^{-(n+2)} \\
    p(2,0,1) &= \frac12 p(0,0,2) + \frac12 p(1,0,2) = \frac{1}{16} \\
    p(2,1,0) &= \frac12 p(0,0,2) = \frac{1}{16} \\
    p(n,0,1) &= \frac12 p(0,0,n) + \frac12 p(1,0,n) = \frac12 p(0,0,n) = 2^{-(n+2)} \\
    p(n,1,0) &= \frac12 p(0,0,n) + \frac12 p(0,1,n) = \frac12 p(0,0,n) = 2^{-(n+2)} \\
\end{aligned}$$

```python
p001 = 0
p002 = 1/8
p00n(n) = 2^(-(n+1))
p010 = 1/8
p020 = 1/16
p0n0(n) = 2^(-(n+2))
p101 = 1/8
p102 = 0
p10n(n) = 0
p110 = 0
p120 = 1/16
p1n0(n) = 2^(-(n+2))
p201 = 1/16
p210 = 1/16
pn01(n) = 2^(-(n+2))
pn10(n) = 2^(-(n+2))
```

Making sure the probabilities add up to one.

```python
(p001+p002+p010+p020+p101+p102+p110+p120+p201+p210
 +sum(p00n(x)+p0n0(x)+p10n(x)+p1n0(x)+pn01(x)+pn10(x),x,3,oo))
```

And the average longest streak is 5/2.

```python
(p001+2*p002+p010+2*p020+p101+2*p102+p110+2*p120+2*p201+2*p210
 +sum(x*(p00n(x)+p0n0(x)+p10n(x)+p1n0(x)+pn01(x)+pn10(x)),x,3,oo))
```
