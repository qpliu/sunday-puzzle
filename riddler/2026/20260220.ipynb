Fiddler
-------
If they both choose the more distant target, the probability of winning is 49%.

If they both choose one of the nearer targets, the probability that each one hits
their chosen target and they both choose the same target is 1/2⋅98/100⋅98/100,
or 98% of 49%, which is less than 49%, so they'll both choose the more distant
target and their probability of winning is 49%.

Extra credit
------------
I'll assume the probability of hitting a target not aimed at is zero.

Let the probability of hitting a nearer target be $C$ and the probability of hitting
the more distant target be $D$.

If they both aim at the more distant target, the probability of winning is $D^2$.

If they both aim at one of the nearer targets, the probability of winning is $C^2/2$.

If $D^2 > C^2/2$, they'll both aim at the more distant target.

If $D^2 < C^2/2$, they'll both aim at one of the nearer targets.

In the remaining case of $D^2 = C^2/2$, the only solution that I can see for which
both $C$ and $D$ are rational is if they are both zero, so their probability of winning
is zero.

### Further thoughts
But if $C$ and $D$ could be irrational, then what would be probability of winning be
if $D^2 = C^2/2$?  One of them might think that aiming at the more distant target makes
sense, since that means not having to choose between the other two, but what if the other
comes to a different conclusion?

If each one has 1/3 probability of choosing each target, then the probability of winning
would be $D^2/9 + 2C^2/9 = 5D^2/9$.
