Fiddler
-------
Since the first of every month after March has the same day of week relative to the
day of week of the first of March in every year, and every day of week is represented
in those ten months, there is no day of week that isn't on the first of a month in any
year.

The special year is 1752, for which Thursday is not the first of any month.

Extra credit
------------
The possible number of days in a month modulo 7 are 0, 1, 2, 3.  They can't all be 0,
because there aren't enough months, so there have to be either two months of 3 (31) for
every month of 1 (29), or two months of 2 (30) for every month of 3 (31).  That means
months must start on at least three different days of the week.

December can be used as a sink, deferring a different day of the week to the following
year.

For years with 365 days, there are three ways for the first of every month to be on no
more than three different days of the week:
  * 31 31 29 31 31 29 31 31 29 31 31 30
  * 31 30 30 31 30 30 31 30 30 31 30 31
  * 30 31 30 30 31 30 30 31 30 30 31 31

For years with 366 days, there is one way for the first of every month to be on no
more than three different days of the week:
  * 31 31 29 31 31 29 31 31 29 31 31 31
