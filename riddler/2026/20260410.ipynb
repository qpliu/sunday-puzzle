Fiddler
-------
To get 001-099, then each of 0-8 need to be on at least two cubes, which is also
18 of the 24 faces.  That leaves 6 faces, which could cover 1-6, meaning 777 would
not be possible, so an upper bound for $N$ is 776.

And 001-776 can be made by 123456, 123678, 045678, and 012345, which I found by
playing around, so $N = 776$.

```python
def get_digits(n):
    h = n // 100
    t = (n // 10) % 10
    o = n % 10
    if o < t or t < h or o == 9 or t == 9 or h == 9:
        return []
    return [o,t,h]

def can_make(digits, cubes):
    if len(digits) == 0:
        return True
    for i in range(len(cubes)):
            if digits[0] in cubes[i]:
                remaining_cubes = cubes.copy()
                del remaining_cubes[i]
                if can_make(digits[1:], remaining_cubes):
                    return True
    return False

def find_N(cubes):
    for i in [1..999]:
        if not can_make(get_digits(i), cubes):
            return i-1
    return None
```

Extra credit
------------
To be able to make 001-776, the 24 faces need to have three of each of 1, 2, 3, 4, 5, 6,
and two of each of 0, 7, 8.

I'll just brute-force all the ways, then filter out non-distinct ways
and ways that can't make 001-776.  And the result is 855 distinct ways.

```python
def add_digit(cubes, digit, count):
    s = (digit,)
    if count == 2:
        if len(cubes[0]) < 6 and len(cubes[1]) < 6:
            yield (cubes[0] + s, cubes[1] + s, cubes[2],     cubes[3])
        if len(cubes[0]) < 6 and len(cubes[2]) < 6:
            yield (cubes[0] + s, cubes[1],     cubes[2] + s, cubes[3])
        if len(cubes[0]) < 6 and len(cubes[3]) < 6:
            yield (cubes[0] + s, cubes[1],     cubes[2],     cubes[3] + s)
        if len(cubes[1]) < 6 and len(cubes[2]) < 6:
            yield (cubes[0],     cubes[1] + s, cubes[2] + s, cubes[3])
        if len(cubes[1]) < 6 and len(cubes[3]) < 6:
            yield (cubes[0],     cubes[1] + s, cubes[2],     cubes[3] + s)
        if len(cubes[2]) < 6 and len(cubes[3]) < 6:
            yield (cubes[0],     cubes[1],     cubes[2] + s, cubes[3] + s)
    else:
        if len(cubes[0]) < 6 and len(cubes[1]) < 6 and len(cubes[2]) < 6:
            yield (cubes[0] + s, cubes[1] + s, cubes[2] + s, cubes[3])
        if len(cubes[0]) < 6 and len(cubes[1]) < 6 and len(cubes[3]) < 6:
            yield (cubes[0] + s, cubes[1] + s, cubes[2],     cubes[3] + s)
        if len(cubes[0]) < 6 and len(cubes[2]) < 6 and len(cubes[3]) < 6:
            yield (cubes[0] + s, cubes[1],     cubes[2] + s, cubes[3] + s)
        if len(cubes[1]) < 6 and len(cubes[2]) < 6 and len(cubes[3]) < 6:
            yield (cubes[0],     cubes[1] + s, cubes[2] + s, cubes[3] + s)

def make_cubes():
    for c0 in add_digit(((),(),(),()),0,2):
        for c1 in add_digit(c0,1,3):
            for c2 in add_digit(c1,2,3):
                for c3 in add_digit(c2,3,3):
                    for c4 in add_digit(c3,4,3):
                        for c5 in add_digit(c4,5,3):
                            for c6 in add_digit(c5,6,3):
                                for c7 in add_digit(c6,7,2):
                                    for c8 in add_digit(c7,8,2):
                                        c8 = list(c8)
                                        c8.sort()
                                        yield tuple(c8)
```

```python
l = [cubes for cubes in set(make_cubes()) if find_N(list(cubes)) >= 776]
len(l)
```

```python
def ts(cube):
    ds = list(cube)
    ds.sort()
    return f"{ds[0]}{ds[1]}{ds[2]}{ds[3]}{ds[4]}{ds[5]}"

def tss(cubes):
    t = [ts(cubes[0]),ts(cubes[1]),ts(cubes[2]),ts(cubes[3])]
    t.sort()
    return f"{t[0]},{t[1]},{t[2]},{t[3]}"
```

```python
t = [tss(cubes) for cubes in l]
t.sort()
t
```
