Fiddler
-------
The three smallest values of $L$ are $1$, $\sqrt2$, $\sqrt5$.  If $L=1$ or if $L=\sqrt2$,
the only possible paths are square loops.

If $L=\sqrt5$, the only other possible path is a rhombus, so the minimum $L$ is $\sqrt5$.

### Post-Update
When there is only a single jump direction, excluding horizontal and vertical reflections and 90° rotations,
the loops can only be squares or rhombuses with a vertex that's a rook's move or a bishop's move away.

The smallest value of $L$ with multiple jump directions is $5$, where the jump directions are $(5,0)$ and
$(3,4)$, and the $(5,0)$ means one of the squares is a rook's move away.

The next value of $L$ is $\sqrt{50}$, with $(7,1)$ and $(5,5)$, and the $(5,5)$ means one of the squares
is a bishop's move away.

The next value of $L$ is $\sqrt{65}$, with $(8,1)$ and $(7,4)$, which has a rhombus loop:

    1............
    ........2....
    .............
    .............
    .............
    .............
    .............
    ....4........
    ............3

So the mimimum $L$ is $\sqrt{65}$.

```python
def dests(start,jumps):
    result = set()
    for j in jumps:
        result.add((start[0]+j[0],start[1]+j[1]))
        result.add((start[0]-j[0],start[1]+j[1]))
        result.add((start[0]+j[0],start[1]-j[1]))
        result.add((start[0]-j[0],start[1]-j[1]))
        result.add((start[0]+j[1],start[1]+j[0]))
        result.add((start[0]-j[1],start[1]+j[0]))
        result.add((start[0]+j[1],start[1]-j[0]))
        result.add((start[0]-j[1],start[1]-j[0]))
    return result

def loops(jumps):
    s1 = (0,0)
    s2 = jumps[0]
    s3_from2 = dests(s2, jumps)
    for s4 in dests(s1,jumps[1:]):
        s3_from4 = dests(s4, jumps)
        for s3 in s3_from2 & s3_from4:
            yield((s1,s2,s3,s4))
```

```python
list(loops([(8,1),(7,4)]))
```

Extra credit
------------
The possible jumps are chess knight moves.

On a 1×1 board, it's not possible to jump to the square.

On a 2×2 board, it's not possible to jump to any square.

On a 3×3 board, when starting from an edge square, it's possible to visit all the edge
squares, but it is not possible to jump to the middle square.  When starting from the
middle square, it's not possible to jump to any square.

On a 4×4 board, it's possible to visit every square of a 3×3 ring that the starting
square is part of and 7 of those 8 squares are also part of at least one other 3×3 ring,
all of which include the middle of the first 3×3 ring, so the minimum $N$ is 4.

### Post-Update
Using code, the minimum $N$ is 14.

```python
def visit(n,jumps):
    visited = set()
    queue = [(1,1)]
    while len(queue) > 0:
        square = queue[0]
        del queue[0]
        if square[0] < 1 or square[0] > n or square[1] < 1 or square[1] > n or square in visited:
            continue
        visited.add(square)
        for s in dests(square, jumps):
            if s[0] >= 1 and s[0] <= n and s[1] >= 1 and s[1] <= n and s not in visited:
                queue.append(s)
    return visited
```
