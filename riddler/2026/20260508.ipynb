Fiddler
-------
Let $x$, $y$, and $z$ be the random amounts scaled to be from 0 to 1.
Doing the integrals yields the expected 7.

```python
x,y,z = var("x,y,z")
r = integral(
        integral(
            y + integral(z, z, 0, x-y) + integral(x-y, z, x-y, 1),
            y,0,x)
        + integral(
            x + integral(z, z, 0, y-x) + integral(y-x, z, y-x, 1),
            y,x,1),
        x,0,1)*12
r
```
