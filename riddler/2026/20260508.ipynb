Fiddler
-------
Let $x$, $y$, and $z$ be the random amounts scaled to be from 0 to 1.
Doing the integrals yields the expected 14.

```python
x,y,z = var("x,y,z")
r = integral(
        integral(
            2*y + integral(2*z, z, 0, x-y) + integral(2*(x-y), z, x-y, 1),
            y,0,x)
        + integral(
            2*x + integral(2*z, z, 0, y-x) + integral(2*(y-x), z, y-x, 1),
            y,x,1),
        x,0,1)*12
r
```
