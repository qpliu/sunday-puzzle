Fiddler
-------
The fastest paths that visit all three points are O-A-O-B-C, O-B-C-O-A, O-C-B-O-A.

If he hides at A, it takes 2 minutes, 14 minutes, and 14 minutes to find him.

If he hides at B, it takes 7 minutes, 3 minutes, and 9 minutes to find him.

If he hides at C, it takes 12 minutes, 8 minutes, and 4 minutes to find him.

Only the first path avoids the possibility of taking 14 minutes, so if he is maximally clever,
it takes 12 minutes to find him.

Extra credit
------------
He'll certainly be at point B from 2 minutes in to 3 minutes in, since 3 minutes in
is the soonest I can be at point B, and 2 minutes is the soonest I can be at point A.

I didn't really understand the problem.  If he could be maximally clever and predict
my every decision, he could stay hidden forever.

The answer seems to be to randomly be at either point A or point B at a given time.
The first possible such time is 3 minutes in, by either walking to point A and waiting
a minute, or by walking to point B.  The next such time is 5 minutes later, by either
waiting 5 minutes or by walking to the other point.  At each of these times, there is
a 50% chance of finding him, so the average time is $3 + 5/2 + 5/4 + 5/8 + \cdots = 8$
minutes.
