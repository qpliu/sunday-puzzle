Fiddler
-------
The two obvious paths are the helix and across the face and down the side.
Let $R = 2\text{m}$ and $h = 2\text{m}$ be the radius and height of the
cylinder.  The length of the latter path is $2R + h = 6\text{m}$.

For the helical path, let $\theta = \pi z/h$, $x = R\cos\theta$, $y = R\sin\theta$.

Then, $dx = -R\pi dz\sin(\pi z/h)/h$, $dy = R\pi dz\cos(\pi z/h)/h$,
$dL = \sqrt{dx^2 + dy^2 + dz^2} = dz \sqrt{1 + R^2\pi^2/h^2}$.

Integrating $z$ from $0$ to $h$, $L = \sqrt{h^2 + \pi^2 R^2}$.

The length of the helical path is $2\sqrt{1+\pi^2} \approx 6.594\text{m}$.

So the shortest path is $6\text{m}$, either going across the face, then down the
side or down the side, then across the face.

### Further thoughts
Let $h = fR$, and let $R = 1$.  The lengths of the two paths are $\sqrt{f^2+\pi^2}$ and $f + 2$.

The two lengths are equal when $f^2 + \pi^2 = f^2 + 4f + 4$, or $f = \pi^2/4 - 1 \approx 1.467$.

When $f < \pi^2/4 - 1$, when the cylinder is more disk-like, the path across the face is shorter,
and when $f > \pi^2/4 - 1$, when the cylinder is more rod-like, the helical path is shorter.

The Fiddler case has $f = 1 < \pi^2/4 - 1$, so the path across the face is shorter.

Extra credit
------------
Let $r = 1\text{m}$ be the inner radius, $R = 2\text{m}$ be the outer radius,
and $h = 2\text{m}$ be the height.

The length of the outer helical path is
$\sqrt{h^2 + \pi^2R^2} \approx 6.594 \text{m}$.

The length of the path going straight to the hollow, going along the around the
perimeter of the hollow, and, at some point, going straight down, is
$2(R-r) + \pi r + h \approx 7.142 \text{m}$.

The length of the path going straight to the hollow, taking the helical path
down the hollow, then straight across the face is
$2(R-r) + \sqrt{h^2 + \pi^2r^2} \approx 5.724 \text{m}$.

The length of the path across the face tangent to the hollow, then going around
the perimeter of the hollow, then across the face to the outer edge, and, at some
point, going straight down, is
$2\sqrt{R^2-r^2} + r(\pi-2\arccos(r/R)) + h \approx 6.511 \text{m}$.

The length of the path across the face tangent to the hollow, then helically down
the hollow, then across the face to the outer edge, is
$2\sqrt{R^2-r^2} + \sqrt{h^2 + r^2(\pi-2\arccos(r/R))^2} \approx 5.722 \text{m}$,
which should be the shortest path.  The exact value is $2\sqrt3 + \sqrt{\pi^2+36}/3 \text{m}$.
