Fiddler

I think they always win.

If the first sees two white hats, they win.

Otherwise, if the second sees that the third is wearing a white hat, they win.

Otherwise, at this point, the third cannot be wearing a white hat, so they win.

Using code to enumerate all 63 triples where they always win, the
greatest sum is 7.

```python
def always_win(bag):
    hat_combos = [(h1,h2,h3,h4) for h1 in ['R','W','B']
            for h2 in ['R','W','B']
            for h3 in ['R','W','B']
            for h4 in ['R','W','B']
            if valid((h1,h2,h3,h4),bag)]
    for turn in range(4):
        hat_combos = remove_wins(hat_combos, turn)
    return len(hat_combos) == 0

def valid(hats,bag):
    (h1,h2,h3,h4) = hats
    (r,w,b) = bag
    if h1 == 'R':
        r -= 1
    if h2 == 'R':
        r -= 1
    if h3 == 'R':
        r -= 1
    if h4 == 'R':
        r -= 1
    if h1 == 'W':
        w -= 1
    if h2 == 'W':
        w -= 1
    if h3 == 'W':
        w -= 1
    if h4 == 'W':
        w -= 1
    if h1 == 'B':
        b -= 1
    if h2 == 'B':
        b -= 1
    if h3 == 'B':
        b -= 1
    if h4 == 'B':
        b -= 1
    return r >= 0 and w >= 0 and b >= 0

def seen(hats,turn):
    (h1,h2,h3,h4) = hats
    if turn == 0:
        return (h2,h3,h4)
    elif turn == 1:
        return (h1,h3,h4)
    elif turn == 2:
        return (h1,h2,h4)
    elif turn == 3:
        return (h1,h2,h3)

def remove_wins(hat_combos,turn):
    possibles = dict()
    for hats in hat_combos:
        k = seen(hats,turn)
        if k not in possibles:
            possibles[k] = set([hats[turn]])
        else:
            possibles[k].add(hats[turn])
    return [hats for hats in hat_combos if len(possibles[seen(hats,turn)]) != 1]
```

```python
[(r,w,b) for r in [0..4]
 for w in [0..4]
 for b in [0..4]
 if r+w+b >= 4 and always_win((r,w,b))]
```
