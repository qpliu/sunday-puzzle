Fiddler

Given $a$, the probability distribution of $r$, a generated random value, is proportional
to $1/a$ for $0 < r < a$, and zero when $r$ is outside that range.

So the probability distribution of $(a,r)$ is

```python
P(a,r) = heaviside(a)*heaviside(1-a)*heaviside(r)*heaviside(a-r)/a
```

```python
assume(0 < a < 1)
integral(integral(P(a,r), r, 0, a), a, 0, 1)
```

Given that $v$ has been generated, $a$ must be greater than or equal to $v$,
so the probability distribution of $a$ is $1/(1-v)$ when $v < a < 1$, and zero
when $a$ is outside that range.

```python
v = var("v")
assume(0 < v < 0.5)
integral(integral(1/a, r, 0, a), a, v, 1)
```

On average, $\langle a\rangle = (v+1)/2$, so when $v = 0.5$, $\langle a\rangle = 0.75$.

```python
integral(a/(1-v),a,v,1).simplify_full(), integral(a/(1-v),a,v,1).subs(v=0.5)
```

My original probability distribution function, $P(a,r)$ is still correct.

If $v$ is generated, then the probability distribution of $a$ must be
proportional to $P(a,v)$, so the average value of $a$ must be
$\int da\,aP/\int da\,P$.

```python
assume(0 < v < 0.5)
Af = integral(a*1/a,a,v,1)/integral(1/a,a,v,1)
Af, Af.subs(v = 1/2).simplify(), numerical_approx(Af.subs(v = 1/2))
```

```python
Ax = (integral(integral(a*1/a, a, v, 1), v, 0, 1/2)
      /integral(integral(1/a, a, v, 1), v, 0, 1/2))
Ax, Ax.simplify_full(), numerical_approx(Ax)
```
